Load the CSV and necessary libraries.

```python
import csv
import numpy as np
import pandas as pd

df = pd.read_csv('dsm-beuth-edl-demodata-dirty.csv')
```

Drop rows where all columns are NAN:

```python
df = df.dropna(how='all')
```

Deduplicate rows that have the same fullname and email:

```python
df = df.drop_duplicates(subset=['full_name','email'])
```

Recreate the id column to ensure all rows have ids:

```python
df.id = np.arange(df['id'].iloc[0], len(df['id'])+1, 1.0)
df
```

Replace NaN and negative ages with the median age:

```python
df.age = pd.to_numeric(df['age'], errors='coerce')
df.loc[(df['age'] < 0) | (df['age'].isnull()), 'age'] = df.age.median()
df
```

Replace NaN genders with the gender that appears the most:

```python
df = df.fillna(df['gender'].value_counts().index[0])
df
```
